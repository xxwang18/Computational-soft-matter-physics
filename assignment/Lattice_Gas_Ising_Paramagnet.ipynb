Problem 1: Lattice Gas (Ising Paramagnet)

A lattice gas consists of particles moving on lattice sites. These sites can either be filled by
one particle, or they can be empty. This system is equivalent to an Ising paramagnet consisting
of N non-interacting spins. Each spin can take the values si ∈ {−1, +1}. The energy H in a
magnetic field H is given by

$H = \mu \sum_{i=1}^N s_i$

The magnetic field H is equivalent to the chemical potential µ of the lattice gas.
Write a Monte Carlo program to simulate this system with the “simple sampling” technique.

(a) Start with an ordered configuration, si = +1 for i = 1, ..., N.

(b) Calculate the time-dependent magnetization $M(t) = \sum_{i=1}^N s_i$
for N = 50, H = 0, and t ≤ 1000.

(c) Calculate the average magnetization
$⟨M⟩ = \frac{1}{Z} \sum_t M(t) exp(- \beta H)$
with normalization factor $Z =\sum_t exp(−βH)$ in thermal equilibrium for N = 50 and
β$\mu$ = 0, 0.1, 0.5, 1.0, 2.0, 5.0. Hint: Use 100 ≤ t ≤ 10000.

(d) Which problems occur for large βH?

```python
"""
Define a function to perform the simulation of Ising model to calculate the magnetization of this system by simple sampling technique

parameters:
N: number of sites
betamu: the value of \beta mu
simtime: The total simulation time

steps:
(1) Start with an ordered configuration, si = +1 for i = 1, ..., N. sumM = 0 (sum of magnetization), Z = 0(normalization factor)  
(2) choose a random spin i and flip the spin si->-si
(3) calculate the magnetization value M(t) = \sum si,and betaH = \beta\mu \sum si,sumM+=M(t)exp(-beta H),Z+=exp(-beta H). return to the step (b) until t=simtime
(4) calculate the average M, <M> = sumM/Z from the range 100-simtime
"""
def simple_sampling(N,betamu,simtime):
    config=np.ones(N)
    sumM = 0 
    Z = 0
    Mt_array = []
    
    for t in range(simtime):
        index = np.random.randint(0, N)
        config[index] = -config[index]
        
        Mt = config.sum()
        Mt_array.append(Mt)
        betaH = betamu*Mt
        if t>100:
            sumM+=Mt*np.exp(-betaH)
            Z+=np.exp(-betaH)
    
    averM = sumM/Z
    return Mt_array,averM
```

(b) Calculate the time-dependent magnetization $M(t) = \sum_{i=1}^N s_i$
for N = 50, $\mu$ = 0, and t ≤ 1000.

```python
N = 50
betamu = 0
simtime =1000

Mt_array,averM = simple_sampling(N,betamu,simtime)

def plot_M(Mt_array,averM,N,betamu,simtime):
    plt.plot(Mt_array,label='M(t)')
    t = np.linspace(0,simtime)
    y = np.full_like(t, averM)
    plt.plot(t,y,label="<M> = "+str(averM))
    plt.xlabel("time")
    plt.ylabel("Manetization")
    plt.title("Time-dependent magnetization M(t) for N = "+str(N)+r" $\beta \mu = $"+str(betamu)+" and  simtime ="+str(simtime))
    plt.legend()
plot_M(Mt_array,averM,N,betamu,simtime)
```

(c) Calculate the average magnetization
$⟨M⟩ = \frac{1}{Z} \sum_t M(t) exp(- \beta H)$
with normalization factor $Z =\sum_t exp(−βH)$ in thermal equilibrium for N = 50 and
β$\mu$ = 0, 0.1, 0.5, 1.0, 2.0, 5.0. Hint: Use 100 ≤ t ≤ 10000.

```python
N=50
simtime=10000
betamu_list = [0,0.1,0.5,1.0,2.0,5.0]
betamu = betamu_list[0]
Mt_array,averM = simple_sampling(N,betamu,simtime)
plot_M(Mt_array,averM,N,betamu,simtime)
    
```
